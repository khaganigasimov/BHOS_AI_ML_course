<h3>Summarization of Methods</h3>

select_dtypes - create a new dataframe from original dataframe so that it will have only chosen data types

loc - create a new dataframe from original dataframe based on conditions

isna() - create a new dataframe having only True of False values indicating whether or not NaN values exist

corr() - create a new dataframe showing correlations between all columns

unique() - return the array that has unique values of the chosen column

boxplot(), bar() - plot the corresponding graphs 

<h3>Applications</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Choose right string for filepath based on the dataset location
filepath = 'Periodic Table of Elements.csv'
df = pd.read_csv(filepath)
df.head()
```

```python
df.shape
```

<h6>Extraction of subdataframe based on special data types</h6>

```python
subdf = df.select_dtypes([np.float64])
```

```python
subdf.shape
```

<h6>Choosing dataframe based on conditions</h6>

```python
conditions = df['Year'] > 2000
conditions
```

```python
df.loc[conditions]
```

<h6>Analyzing Nan values</h6>

```python
## Boolean dataframe indicating whether or not Nan Values exist
df.isna()
```

```python
## The number of Nan values per column
df.isna().sum()
```

<h6>Obtaining correlations between all columns</h6>

```python
## The reason behind warning is that the dataframe contains non-numerical data which can't be used for mathematical operations.
## To avoid this, simply you can choose such a subdataframe that has only float or integer objects (Use select_dtypes)
df.corr()
```

<h6>Obtaining unique values</h6>

```python
df['Period'].unique()
```

<h6>Obtaining index and values from groupped dataframe</h6>

```python
groupped_period = df.groupby('Period')
group_based_on_density = groupped_period['Density'].mean()
group_based_on_density
```

```python
index = group_based_on_density.index
values = group_based_on_density.values
```

```python
index
```

```python
values
```

<h6>Plotting bar and box charts</h6>

```python
## 1 and 2 indicates the number of rows and columns, respectively

fig, ax = plt.subplots(1, 2, figsize = (7, 4))

ax[0].bar(index, values)
ax[1].boxplot(df['AtomicMass'])
```
