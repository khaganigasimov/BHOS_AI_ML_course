<h5>The aim is to explain how parameters restrict the tree from developing</h5>

Maximum depth indicates the maximum allowable number of depths. Imagine you have a tree having the depth of 10 by default. As we know, the tree always tries to minimize errors on the given dataset. To prevent potential overfitting, we can define max_depth, say as 5 so that the tree would stop growing in case of reaching to 5. However, it doesn't mean tree must reach to 5. If the max_depth is set as 15, the tree probably stops in the depth of 10.

Consider max_depth = 3. This time, we defines min_samples_split as 300. This parameter is another restriction for the tree which  helps stop growing in the case that number of samples in the node is lower than min_samples_split. If this condition is valid, that node turns out to become a leaf.
